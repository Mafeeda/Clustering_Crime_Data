## Normalizing Data

```python
#from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime_data = scaler.fit_transform(crime_data.iloc[:,1:])
```

```python
scaled_crime_data
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime_data)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, random_state = 42)
kmeans = kmeans.fit(scaled_crime_data)

```

```python
kmeans.labels_
```

```python
# Adding cluster to our data set
crime_data['cluster'] = kmeans.labels_
```

```python
#Mean of clusters 1 to 4
kmeans_mean_cluster = pd.DataFrame(round(crime_data.groupby('cluster').mean(),1))
kmeans_mean_cluster
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=crime_data['Murder'], y = crime_data['Assault'],hue = kmeans)
```
